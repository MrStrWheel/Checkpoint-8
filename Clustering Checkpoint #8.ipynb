```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("CreditCards.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
data["MINIMUM_PAYMENTS"].fillna(data["MINIMUM_PAYMENTS"].mean(), inplace=True)
```

```python
data.isnull().sum()
```

```python
data.dropna(axis=0, how='any', thresh=None, inplace=True)
```

```python
data.isnull().sum()
```

```python
def data_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 20 , 20 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
data_correlation_map(data)
```

```python
#After giving it some thought, i've come to the conclusion that our goal is to divide the customers into 5 classes:
#extravagants, cautious, average, generous, and cheap.
#this means that the most important features we're gonna use are how much money they got and how much they spend.
#in our case, these two features are Balance and Purchases.
#let's prepare our data:
datacl = data[["BALANCE","PURCHASES"]]
```

```python
#let's try out hierarchial clustering:
from sklearn.cluster import AgglomerativeClustering
AggClust=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=AggClust.fit_predict(datacl)
```

```python
#let's label our data:
lbl=pd.DataFrame(clust_labels)
lbl
```

```python
#let's now visualize the clusters:
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (data ['BALANCE'] , data ["PURCHASES"] , c= lbl[0], s=50)
ax.set_title("Clusters")
ax.set_xlabel("Balance")
ax.set_ylabel("Purchases")
plt.colorbar(scatter)
```

```python
#customers in purple are cautious customers, which are the majority.
#customers in blue are cheap customers, which are also pretty dominant.
#customers in dark green/light blue are normal customers.
#customers in light green are extravagant customers.
#customers in yellow are generous customers.
#let's now plot the dendogram.
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title("Customer Dendrograms")
dend=shc.dendrogram(shc.linkage(datacl, method="complete"))
```

```python
from sklearn.cluster import KMeans  
kmeans=KMeans(n_clusters=5, random_state=0)
kmeans.fit(datacl)
```

```python
labels=pd.DataFrame(kmeans.labels_)
labels
```

```python
kmeans.predict(datacl)
print(kmeans.cluster_centers_) 
```

```python
plt.scatter(data["BALANCE"],          
            data["PURHCASES"],s=80,c='magenta',label='Careful')
plt.scatter(data["BALANCE"],
           data["PURHCASES"],s=80,c='yellow',label='Standard')
plt.scatter(data["BALANCE"],
           data["PURHCASES"],s=80,c='green',label='Target')
plt.scatter(data["BALANCE"], 
           data["PURHCASES"],s=80,c='cyan',label='Careless')
plt.scatter(data["BALANCE"], 
           data["PURHCASES"],s=80,c='burlywood',label='Sensible')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title('Clusters of Customers')
plt.xlabel('Balance')
plt.ylabel('Purchases')
plt.legend()
plt.show()
```
